## Clean

```python
df_c = df.copy()
```

- The first column should be removed
- Not all columns are needed
#### Define
- use masking to get rid of the unwanted columns in one step
#### Code &  Test

```python
df_c = df_c[['course_id','published_title','is_paid','num_subscribers','rating','num_reviews','num_published_lectures','num_published_practice_tests','instructional_level','published_at','content_info','category']]
df_c
```

- You need to extract the year the course had been published
#### Define
- use RegEX
#### Code &  Test

```python
df_c.insert(9,'publish_year', value = "")
df_c['publish_year'] = df_c['published_at'].str.findall("[0-9]{4}").str[0]
df_c
```

- is_paid column can be presented in a better way (name : free/paid, change type to string)
#### Define
- use rename replace the values
#### Code &  Test

```python
df_c['is_paid'] = df_c['is_paid'].astype('str').replace(['True','False'], ['paid','free'])
```

```python
df_c['is_paid'].value_counts()
```

- Sometimes the duration represented in minutes, others in hours
#### Define
- represent them in hours by extracting the unit in a different column and divide the minutes/60
#### Code &  Test

```python
df_c.insert(12,'time_unit', value ="")
df_c['content_info'] = df_c['content_info'].str.replace('total','')
```

```python
df_c
```

```python
df_c['time_unit'] = df_c['content_info'].str.findall("[a-z]+").str[0]
df_c['time_unit'].value_counts()
```

```python
df_c['time_unit'] = df_c['time_unit'].replace(['hour','min'],['hours','mins'])
df_c['time_unit'].value_counts()
```

```python
df_c[df_c['time_unit']=='questions']
```

```python
print(df_c[df_c['time_unit']=='questions']['published_title'])
```

```python
df_c.insert(13,'teaching_method', value ="")
df_c['teaching_method'] = df_c['time_unit'] 
df_c['teaching_method'] = df_c['teaching_method'].replace(['hours','mins','questions'],['lectures','lectures','practical questions'])
df_c['teaching_method'].value_counts()
```

```python
df_c
```

```python
df_c.insert(12,'duration_hours', value ="")
```

```python
df_c['duration_hours'] = df_c['content_info'].str.split(' ',-1).str[0]
df_c
```

```python
df_c.info()
```

```python
df_c['duration_hours'] = df_c['duration_hours'].astype('float')
```

```python
for i in df_c.index:
    if df_c['time_unit'][i] == 'mins':
        df_c['duration_hours'][i] = round(df_c['duration_hours'][i]/60,2)
    elif df_c['time_unit'][i] == 'questions':
        df_c['duration_hours'][i] = 'None'
        
```

```python
df_c
```

 - 17 nulls in 'instructional_level'
#### Define
- see what you can do
#### Code &  Test

```python
df_c[df_c.instructional_level.isnull()]
```

```python
# Since we can't detect the level , we will have to remove them
df_c = df_c[~df_c.instructional_level.isnull()]
df_c.info()
```

```python
df_c[df_c['time_unit'].isnull()]
```

```python
# The zeros in content_info made nulls in the derived columns
# Since we will not use these zeros, we can delete their columns
df_c = df_c[~df_c.teaching_method.isnull()]
df_c.info()
```

- Rename the columns labels to be more descriptive
#### Define
- use rename method
#### Code &  Test

```python
df_c.rename(columns = {'is_paid':'free/paid','num_subscribers':'subscribers_count','num_reviews':'reviews_count','num_published_lectures':'lectures_count','num_published_practice_tests':'practice_tests_count'},inplace = True)
df_c
```

- remove the unnecessary columns
#### Define
- use drop method
#### Code &  Test

```python
df_c.drop(['published_title','published_at','time_unit','content_info'], axis = 1, inplace = True)
df_c
```

- category has some issues (marketting, personal_development, health_fitness)
#### Define
- use replace as an easy option
#### Code &  Test

```python
df_c['category'] = df_c['category'].replace(['marketting','personal_development','health_fitness'], ['marketing','personal development','health & fitness'])
df_c['category'].value_counts()
```

```python
df_c.to_csv('udemy_courses_cleaned.csv', index = False)
```
